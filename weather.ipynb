```python
import numpy as np
import pandas as pd
import pickle
from datetime import datetime
```

```python
kenner = pd.read_pickle('kenner_scrape.pkl')
```

```python
baton_rouge = pd.read_pickle('baton_rouge_scrape.pkl')
```

```python
lake_charles = pd.read_pickle('lake_charles_scrape.pkl')
```

```python
alexandria = pd.read_pickle('alexandria_scrape.pkl')
```

```python
shreveport = pd.read_pickle('shreveport_scrape.pkl')
```

```python
all_cities = pd.concat([kenner, baton_rouge, lake_charles, alexandria, shreveport])
```

```python
all_cities
```

```python
#create new dataframe with desired features
all_cities = all_cities[['ID', 'avg_temp_deg_f', 'avg_dew_humidity_pct', 'avg_wind_speed_mph']]
all_cities
```

```python
all_cities[['avg_temp_deg_f', 'avg_dew_humidity_pct', 'avg_wind_speed_mph']] = all_cities[['avg_temp_deg_f', 'avg_dew_humidity_pct', 'avg_wind_speed_mph']].astype(float)
```

```python
all_cities[["city", "dtime"]] = all_cities.ID.str.split('_', expand = True)
```

```python
del all_cities["ID"]
all_cities
```

```python
average_weather = all_cities[["dtime","avg_temp_deg_f","avg_dew_humidity_pct","avg_wind_speed_mph"]].groupby(["dtime"]).aggregate(np.average)
```

```python
average_weather
```

```python
average_weather.reset_index(inplace = True)
```

```python
#create new feature - temp_deviation - which calculates temp deviation from 65 degrees
average_weather["temp_deviation"] = np.abs(average_weather['avg_temp_deg_f'] - 65)
```

```python
#create new feature - HDD_CDD - which identifies if day is heating or cooling degree day
average_weather['HDD_CDD'] = np.where(average_weather['avg_temp_deg_f']>65, 1, 0)
average_weather
```

```python
average_weather["dtime"] = pd.to_datetime(average_weather['dtime'])
```
